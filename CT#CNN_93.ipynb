```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix

def evaluate_with_misclassified(cfg, encoder, model, dataset, estimator):
    model.eval()
    misclassified = []
    all_labels = []
    all_preds = []

    with torch.no_grad():
        for batch_idx, (data, target) in enumerate(dataset):
            data, target = data.to(cfg.base.device), target.to(cfg.base.device)
            features = encoder(data)
            output = model(features)
            pred = output.argmax(dim=1, keepdim=True)

            # Store misclassified samples
            mis_mask = (pred.view_as(target) != target)
            for i in np.where(mis_mask.cpu().numpy())[0]:
                misclassified.append({
                    'image': data[i].cpu(),
                    'true': target[i].cpu().item(),
                    'pred': pred[i].cpu().item(),
                    'prob': torch.nn.functional.softmax(output[i], dim=0).cpu().numpy()
                })

            all_labels.extend(target.cpu().numpy())
            all_preds.extend(pred.cpu().numpy())

    # Generate confusion matrix
    cm = confusion_matrix(all_labels, all_preds)
    print("Confusion Matrix:")
    print(cm)

    return misclassified
```

```python
def plot_misclassified(misclassified_samples, class_names, num_samples=10):
    plt.figure(figsize=(15, 10))
    for i, sample in enumerate(misclassified_samples[:num_samples]):
        plt.subplot(2, 5, i+1)
        image = sample['image'].permute(1, 2, 0).numpy()
        if image.shape[2] == 1:  # Grayscale
            image = image.squeeze()
            plt.imshow(image, cmap='gray')
        else:  # RGB
            plt.imshow(image)

        plt.title(f'True: {class_names[sample["true"]]}\nPred: {class_names[sample["pred"]]}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```
